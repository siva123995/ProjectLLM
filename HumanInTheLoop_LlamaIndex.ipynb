```python
from llama_index.embeddings.google_genai import GoogleGenAIEmbedding
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader

# 1. Load your local documents from a folder named "data"
documents = SimpleDirectoryReader("data").load_data()

embed_model = GoogleGenAIEmbedding(
    model = "models/embedding-001",
    api_key=GOOGLE_API_KEY
)

# 2. Index the data into a searchable vector store, using a local embedding model
index = VectorStoreIndex.from_documents(documents, embed_model=embed_model)

# 3. Create a query engine from the index
query_engine = index.as_query_engine(llm=model)

# 4. Ask a question about your data
response = query_engine.query("What does the documents talk about Sikh separatist movement")

# 5. Output the answer
print(response)
```
