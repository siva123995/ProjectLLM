```python
import os
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY

agent = Agent(
    role = "You're a Motivational speaker",
    goal="Create an inspirational LinkedIn post",
    backstory = "An experienced career mentor inspires others by writing",
    llm=LLM(provider="google", model="gemini-3.5-flash")
)

task = Task(
    description="Write a short LinkedIn Post to motivate job seekers",
    expected_output="A compelling LinkedIn post aimed at motivating job seekers.",
    agent=agent
)

crew = Crew(
    agents = [agent],
    tasks = [task]
)
```

```python
import nest_asyncio
nest_asyncio.apply()

result = await crew.kickoff_async()
print(result)
```

# Custom Input

```python
import os
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY

agent1 = Agent(
    role = "You're a Motivational speaker",
    goal="Create an inspirational LinkedIn post",
    backstory = "An experienced investor, looking to invest in real estate",
    llm=LLM(provider="google", model="gemini-3.5-flash")
)

task1 = Task(
    description="Write a short LinkedIn Post to motivate Real estate investors agents about the topic: {topic}",
    expected_output="Clean and humanised LinkedIn post",
    agent=agent1
)

crew1 = Crew(
    agents = [agent1],
    tasks = [task1]
)
```

```python
import nest_asyncio
nest_asyncio.apply()

result = await crew1.kickoff_async(inputs={"topic":"Negative gearing removal in Australia, and CGT changes"})
print(result)
```
