```python
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage, AIMessage, BaseMessage
from langchain_community.tools import TavilySearchResults
from typing import TypedDict, List

search_tool = TavilySearchResults(tavily_api_key = TAVILY_API_KEY)

def planning(state):
  user_input = state["messages"][-1].content

  prompt = (
      "You are helpful assistant. User will pass a question. You need to decide if you need to search to answer user question, if yes, prepare the user question by yourself"
  )

  response = model.invoke(
      [
          HumanMessage(content = prompt),
          HumanMessage(content = user_input)
      ]
  )
  return {"messages": state["messages"] + [AIMessage(content = response.content)]}

def call_search_tool(state):
  query = state["messages"][-1].content
  print(f"{query=}")
  search_results = search_tool.invoke(query)
  return {"messages": state["messages"] + [AIMessage(content=str(search_results))]}

def responder(state):
  user_input = state["messages"][0].content
  search_results = state["messages"][-1].content

  prompt = (
      "you're are a helpful responder. use the following search results to respond user question  "
  )

  response = model.invoke(
      [
          HumanMessage(content = prompt),
          HumanMessage(content = f"User Question is {user_input} and search results: {search_results}")
      ]
  )
  return {"messages": state["messages"] + [AIMessage(content = response.content)]}




# Define the state for our graph
class AgentState(TypedDict):
    messages: List[BaseMessage]

graph = StateGraph(AgentState)

# Add nodes to the graph
graph.add_node("planner", planning)
graph.add_node("search", call_search_tool)
graph.add_node("responder", responder)

# Set the entry point
graph.set_entry_point("planner")

# Add edges
graph.add_edge("planner","search")
graph.add_edge("search","responder")
graph.add_edge("responder", END)

# Compile the graph
agent = graph.compile()

# Invoke the agent
response = agent.invoke({"messages": [HumanMessage(content="Tell me the person name who holds orange cap and purple cap in indian permier league 2026??")]})
```

```python
import os
from typing import Annotated, List, TypedDict
# from langchain_openai import ChatOpenAI # Or your preferred model
from langgraph.graph import StateGraph, END, START
from langchain_core.messages import HumanMessage, AIMessage, BaseMessage

from langchain_community.tools import TavilySearchResults
from langgraph.graph.message import add_messages

# Configuration
from google.colab import userdata
TAVILY_API_KEY = userdata.get('TavilyKey')
GOOGLE_API_KEY = userdata.get('Secret')
# Initialize Model and Tool
# model = ChatOpenAI(model="gpt-4o") # Adjust to your model variable
search_tool = TavilySearchResults(max_results=3, tavily_api_key=TAVILY_API_KEY)

# 1. Define State
class AgentState(TypedDict):
    # Annotated handles appending new messages to the list automatically
    messages: Annotated[List[BaseMessage], add_messages]
    # We store the raw search results here so they don't clutter the chat history
    search_query: str
    search_results: str

# 2. Define Nodes
def planning(state: AgentState):
    user_input = state["messages"][-1].content

    prompt = (
        "You are a search planner. Based on the user's request, "
        "write a concise search query to find the most up-to-date information."
    )

    response = model.invoke([
        HumanMessage(content=prompt),
        HumanMessage(content=user_input)
    ])

    # Fix: Ensure search_query is a string
    content = response.content
    if isinstance(content, list):
        query = next((item["text"] for item in content if isinstance(item, dict) and "text" in item), str(content))
    else:
        query = content

    # We save the query to the state but don't add it to messages
    # to keep the conversation history clean for the user.
    return {"search_query": query}

def call_search_tool(state: AgentState):
    query = state["search_query"]
    print(f"{query=}")
    results = search_tool.invoke(query)

    # Save results to the specific state key
    return {"search_results": str(results)}

def responder(state: AgentState):
    # We take the original user question and the search results
    user_question = state["messages"][0].content
    context = state["search_results"]

    prompt = f"""You are a helpful responder.
    Use the following search results to answer the user's question accurately.

    SEARCH RESULTS:
    {context}
    """

    response = model.invoke([
        HumanMessage(content=prompt),
        HumanMessage(content=user_question)
    ])

    # Return the final AIMessage to be added to the conversation
    return {"messages": [response]}

# 3. Build Graph
workflow = StateGraph(AgentState)

workflow.add_node("planner", planning)
workflow.add_node("search", call_search_tool)
workflow.add_node("responder", responder)

workflow.add_edge(START, "planner")
workflow.add_edge("planner", "search")
workflow.add_edge("search", "responder")
workflow.add_edge("responder", END)

agent = workflow.compile()

# 4. Invoke
inputs = {"messages": [HumanMessage(content="Which team is top scorer in Indian Premier league 2026?")]}
result = agent.invoke(inputs)

# 5. Print only the final response
print("\n--- FINAL ANSWER ---")
print(result["messages"][-1].content)

# If you want to see that it "remembered" the whole flow:
# print("\n--- FULL MESSAGE HISTORY ---")
# for msg in result["messages"]:
#    print(f"{type(msg).__name__}: {msg.content[:100]}...")
```
