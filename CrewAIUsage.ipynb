```python
import os
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY

agent1 = Agent(
    role = "You're a Motivational speaker",
    goal="Create an inspirational LinkedIn post",
    backstory = "An experienced investor, looking to invest in real estate",
    llm=LLM(provider="google", model="gemini-3.5-flash")
)

task1 = Task(
    description="Write a short LinkedIn Post to motivate Real estate investors agents about the topic: {topic}",
    expected_output="Clean and humanised LinkedIn post",
    agent=agent1
)

crew1 = Crew(
    agents = [agent1],
    tasks = [task1]
)
```

```python
import os
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
from crewai_tools import FirecrawlSearchTool

search_tool = FirecrawlSearchTool(api_key = FIRECRAWL_API_KEY)

researcher = Agent(
    role = "You're a Current affairs researcher",
    goal="Fetch top news in the current tech world",
    backstory = "An expert news researcher with access to latest information",
    llm=LLM(provider="google", model="gemini-3.5-flash"),
    tools = [search_tool]
)

summariser = Agent(
    role = "Note maker",
    goal="Convert research into smart bullted points",
    backstory = "An expert in simplifying complex topics for fast reading",
    llm=LLM(provider="google", model="gemini-3.5-flash")
)

task1 = Task(
    description="Search top global tech news stories using the web.",
    expected_output="Top tech global news stroies and its details",
    agent=researcher
)

task2 = Task(
    description="Take the news articles from previous tasks and write clear 1-line summaries for each.",
    expected_output="Create a summarised and consumable notes",
    agent=summariser
)

current_tech_affairs = Crew(
    agents = [researcher, summariser],
    tasks = [task1, task2],
    process = "sequential"
)
```

```python
import nest_asyncio
nest_asyncio.apply()

result = await current_tech_affairs.kickoff_async()
print(result)
```

```python
import os
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
from crewai_tools import FirecrawlSearchTool

per_agent = Agent(
    role = "Friendly Motivator",
    goal="Encourage users with personalized messages",
    backstory = "An assistant that learns user names and motivates them kindly",
    llm=LLM(provider="google", model="gemini-3.5-flash"),
    tools = [search_tool]
)

# Human in the loop
user_name = input("Hi! What is your name? ")

# summariser = Agent(
#     role = "Note maker",
#     goal="Convert research into smart bullted points",
#     backstory = "An expert in simplifying complex topics for fast reading",
#     llm=LLM(provider="google", model="gemini-3.5-flash")
# )

task1 = Task(
    description=f"The user's name is {user_name}. Remember it for future messages",
    expected_output="Remember the current username",
    agent=per_agent
)

task2 = Task(
    description="Using the user'name from memory, write a short motivational message.",
    expected_output="Personal motivated message with the given name",
    agent=per_agent
)

per_task = Crew(
    agents = [per_agent],
    tasks = [task1, task2],
    process = "sequential",
    verbose = True
)
```

# Memory

```python
import os
from crewai import Crew, Agent, Task, Process, LLM
from langchain_google_genai import ChatGoogleGenerativeAI,GoogleGenerativeAIEmbeddings

os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
os.environ["GEMINI_API_KEY"] = GOOGLE_API_KEY

from crewai_tools import FirecrawlSearchTool

researcher = Agent(
    role = "Research Analyst",
    goal="Gather Comprehensive information on given topic",
    backstory = "Expert in online research and information synthesis",
    llm=LLM(provider="google", model="gemini-3.5-flash"),
    tools = [search_tool],verbose=True
)

writer = Agent(
    role = "content writer",
    goal="Create a creative content and engaging content based on research",
    backstory = "Skilled in writing articles and reports",
    llm=LLM(provider="google", model="gemini-3.5-flash"),
    verbose = True
)

research_task = Task(
    description=f"Research the latest information in Negative gearing and investment properties in Australia in 2026",
    expected_output="Researched informaiton about the topic ",
    agent=researcher
)

write_task = Task(
    description="Write a concise report based on the research that happened. ",
    expected_output="Concise information about the topic.",
    agent=writer
)

my_crew = Crew(
    agents = [researcher, writer],
    tasks = [research_task , write_task],
    process = "sequential",
    verbose = True,
    memory = True,
    embedder = {
        "provider":"google-generativeai",
        "config" : {
            "api_key" : os.environ["GOOGLE_API_KEY"],
            "model" : "models/embedding-001"
        }
    }
)
```

```python
 import nest_asyncio
nest_asyncio.apply()

result = await my_crew.kickoff_async()
print(result)
```
